# GroupBy Operations

Average salary by department

GroupBy with Multiple Aggregations

```python
df.groupby('department')['salary'].agg(['mean', 'min', 'max'])
```

GroupBy Multiple Columns

Using agg() with Custom Column Mapping

```python
df.groupby('department').agg({
    'salary': ['mean', 'sum'],
    'age': 'mean',
    'employee_id': 'count'
})
```

```python
df.groupby('department').agg(
    avg_salary=('salary', 'mean'),
    total_salary=('salary', 'sum'),
    employee_count=('employee_id', 'count')
)
```

GroupBy + Filtering (HAVING-like behavior)

```python
df.groupby('department')['salary'].mean().loc[lambda x: x > 50000]
```

# apply() and Lambda Functions

Apply on a Single Column

```python
df['salary_increased'] = df['salary'].apply(lambda x: x * 1.10)
df
```

Apply with Conditional Logic

```python
df['salary_level'] = df['salary'].apply(
    lambda x: 'High' if x >= 60000 else 'Medium' if x >= 50000 else 'Low'
)
```

Apply on Rows (axis=1)

```python
df['performance_label'] = df.apply(
    lambda row: 'Excellent' if row['performance_rating'] >= 4 else 'Needs Improvement',
    axis=1
)
df
```

Apply with Multiple Columns

```python
df['bonus'] = df.apply(
    lambda row: row['salary'] * 0.10 if row['performance_rating'] >= 4 else row['salary'] * 0.05,
    axis=1
)
df
```

apply() vs Vectorized Operations

```python
df['salary_increased_fast'] = df['salary'] * 1.10
df
```

Using Named Functions with apply()

```python
def salary_category(salary):
    if salary >= 60000:
        return 'High'
    elif salary >= 50000:
        return 'Medium'
    else:
        return 'Low'

df['salary_category'] = df['salary'].apply(salary_category)
df
```
